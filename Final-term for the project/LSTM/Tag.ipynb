```python
import pandas as pd
import classification
```

```python
data = pd.read_csv('Purchase.csv')
new_data = classification.classify_third_party(data)
new_data.drop(new_data.columns[[5]], axis=1, inplace=True)
new_data.head(10)
```

```python
# 新建一列`tag`，默认值可以设置为一个特定的整数，比如`-1`
new_data['tag'] = -1

# 然后将所有`Third Party Name`为`Bank`的记录的`tag`设置为`0`
new_data.loc[new_data['Third Party Name'] == 'Bank', 'tag'] = 0
new_data.loc[new_data['Third Party Name'] == 'Indoor Entertainment', 'tag'] = 1
new_data.loc[new_data['Third Party Name'] == 'Fashion & Clothing', 'tag'] = 2
new_data.loc[new_data['Third Party Name'] == 'Kids', 'tag'] = 3
new_data.loc[new_data['Third Party Name'] == 'Arts & Craft', 'tag'] = 4
new_data.loc[new_data['Third Party Name'] == 'Grocery', 'tag'] = 5
new_data.loc[new_data['Third Party Name'] == 'Daily', 'tag'] = 6
new_data.loc[new_data['Third Party Name'] == 'Large Retailers', 'tag'] = 7
new_data.loc[new_data['Third Party Name'] == 'Medical', 'tag'] = 8
new_data.loc[new_data['Third Party Name'] == 'Food & Drink', 'tag'] = 9

new_data.to_csv('cleaned.csv', index=False)
```
