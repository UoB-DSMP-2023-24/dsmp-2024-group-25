```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
data = pd.read_csv("rfm-table-lr.csv")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 创建一个基于整数标签的颜色映射
labels = data['RFMClass'].astype('category').cat.codes
cmap = plt.get_cmap('viridis') 
sc = ax.scatter(data['R_Quartile'], data['F_Quartile'], data['M_Quartile'], c=labels, cmap=cmap, alpha=0.7)
ax.set_xticks(np.arange(min(data['R_Quartile']), max(data['R_Quartile'])+1, 1))
ax.set_yticks(np.arange(min(data['F_Quartile']), max(data['F_Quartile'])+1, 1))
ax.set_zticks(np.arange(min(data['M_Quartile']), max(data['M_Quartile'])+1, 1))

# 为每个点的RFM值创建散点图

cbar = plt.colorbar(sc, ax=ax, label='RFM Class', ticks=[])
cbar.set_label('RFM Class', rotation=270, labelpad=15)

ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')

plt.title('3D Scatter Plot of RFM Analysis')
plt.show()
```
